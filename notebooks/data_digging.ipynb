### Get Num Tweets

```python
def get_num_tweets_col(candidate_df, tweet_df):
    num_tweets_map = {}
    for i, r in tweet_df.iterrows():
        if r['twitter_handle'] in num_tweets_map:
            num_tweets_map[r['twitter_handle']] += 1
        else:
            num_tweets_map[r['twitter_handle']] = 1
    
    num_tweets_col = []
    for i, r in candidate_df.iterrows():
        if r['twitter_handle'] in num_tweets_map and r['twitter_handle2'] in num_tweets_map:
            num_tweets_col.append(num_tweets_map[r['twitter_handle']] + num_tweets_map[r['twitter_handle2']])
        elif r['twitter_handle'] in num_tweets_map:
            num_tweets_col.append(num_tweets_map[r['twitter_handle']])
        else:
            num_tweets_col.append(0)
    return num_tweets_map, num_tweets_col
```

```python
_, col = get_num_tweets_col(dem_df, dem_tweets)
dem_df['num_tweets'] = col
# dem_df.to_csv('../data/num_tweets.csv', index=False)
```

### Does most tweets win??

```python
def get_most_tweets(candidates):
    most_tweets = -1
    candidate_most_tweets = ""
    actual_winner = ""
    for c in candidates:
        n = dem_df[dem_df['candidate'] == c]['num_tweets'].values[0]
        if n > most_tweets:
            candidate_most_tweets = c
        if dem_df[dem_df['candidate'] == c]['won_primary'].values[0] == 'Yes':
            actual_winner = c
    return candidate_most_tweets, actual_winner


count = 0
for district in dem_df['district'].unique():
    candidates = dem_df[dem_df['district'] == district]['candidate'].values
    candidate_most_tweets, actual_winner = get_most_tweets(candidates)
    if actual_winner == candidate_most_tweets:
        count += 1
print(count)
print(count / len(dem_df))
```
